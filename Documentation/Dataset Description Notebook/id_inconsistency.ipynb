# ID Inconsistency over time

This dataset is obtained from real data of the French grid. This has the advantage of having a very realistic topology to test the robustness of the algorithms that can be applied to it (optimization of switching, forecasting of loading of lines, grid planning...). However, this means that the different snapshots 'evolve' over time, with some elements being added to the model or dropping off at certain snapshots. 

Part of the ID changes are due to inconsistent IDs, normally due to small internal changes of the naming of the items, and normally occur in switches that connect exactly the same element to the same busbar. The questions and  can consult some of these kind of inconsistencies in the [Issues](https://github.com/OpenSynth-energy/D-GITT/issues) of the repository. If the switches form part of the same path as in previous snapshots, connecting the same elements, then its inconsistent ID is matched with the previous ID.

This is far more difficult with other grid elements such as transformers, generators, or loads. Right now, there is no fix applied to those elements, since what may seem as a small change in its ID, can refer to a completely diferent unit in the distribution side.

The present notebook pretends to be informative by presenting the statistics obtained for the addition or deletion of IDs between consecutive snapshots. The statistics shown only take into account one snapshot per day, as it is enough to depict this changing nature of the dataset. In particular, we cover here from January 1st 2021 to May 20th 2021, for a total of 140 days.
